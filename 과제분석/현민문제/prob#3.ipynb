```python
# 데이터 int형 변환
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/stroke_/train.csv')
df['age'] = df['age'].str.replace('*','').astype(int)
```

```python
# 결측치 제외 나이대별 평균값 계산 및 dict형태로 변환|
mean = df[df.age.notna()].groupby(df.age // 10 * 10).bmi.mean()
mean
```

```python
dic = {x:y for x, y in mean.items()}
```

```python
# 결측치 매핑
df.loc[df.bmi.isna(), 'bmi'] = (df[df.bmi.isna()].age//10 * 10).map(lambda x: dic[x])
```

```python
result = df.bmi.mean()
print(result)
```

```python
import pandas as pd
train= pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/stroke_/train.csv')
test= pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/stroke_/test.csv')

display(train.head())
display(test.head())
```

```python
train.isnull().sum()
```

```python
test.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_auc_score

train.age = train.age.str.replace('*', '')
train.bmi = train.bmi.fillna(value=train.bmi.mean())
test.bmi = test.bmi.fillna(value=test.bmi.mean())
x = train.drop(columns=['id','stroke'])
x_d = pd.get_dummies(x)
y = train['stroke']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_d,y,stratify=y, random_state=1)
print(x_train.shape)
```

```python
model = RandomForestClassifier()
model.fit(x_train,y_train)
pred = model.predict_proba(x_test)
```

```python
print(f'test acc is {roc_auc_score(y_test,pred[:,1])}')
```
